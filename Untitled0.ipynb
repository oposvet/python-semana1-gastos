```python
# 1. Pide al usuario su nombre (usa input())
nombre = input("¿Cuál es tu nombre? ")

# 2. Pide el gasto semanal en números (convierte a float con float())
gasto = float(input("¿Cuánto gastaste esta semana en €? "))

# 3. Calcula si se pasó del presupuesto (100€)
presupuesto = 100.0
diferencia = presupuesto - gasto

# 4. Muestra mensaje personalizado
if diferencia >= 0:
    print(f"¡Buen trabajo {nombre}! Te sobran {diferencia:.2f}€")
else:
    print(f"{nombre}, te pasaste por {-diferencia:.2f}€. ¡Ajusta la próxima!")

# Resultado esperado EJEMPLO:
# Input: "Juan", "120"
# Output: "Juan, te pasaste por 20.00€. ¡Ajusta la próxima!"
```
